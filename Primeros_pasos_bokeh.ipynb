```python
# se importan librerias de bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook
```

```python
#se crean datos normalitos
x = [0,1,2,3,4,5]
y = [5,6,7,8,9,10]
```

```python
#archivo de salido que se guardara en tu carpeta para interactuar con el grafico
output_file("line.html")
# se crea la figura como en matplotlib
f=figure()
# se rellena la figura
f.line(x,y)

f.circle(x,y)

f.triangle(x,y)

# muestra la figura creada...
show(f)
```

```python
#metodos de que figure tiene
dir(f)
```

# Bokeh introduccion

```python
from bokeh.sampledata.iris import flowers
```

```python
# crea la figura 
f = figure()
# aqui podemos irla modificando
f.plot_width = 1100 #ancho (en pixeles)
f.plot_height = 800 #largo (en pixeles)
f.background_fill_color="olive" #color del fondo
f.background_fill_alpha=0.6 #trasnparencia del fondo

# titulo
f.title.text = 'Iris lenght and weight'
f.title.text_color = 'olive'
f.title.text_font = 'times'
f.title.align = 'center'
f.title.text_font_size = '24px'

#axis
f.xaxis 
```

```python
# grafico de dispersion e iremos mejorandolo
f.circle(x=flowers.petal_length, y=flowers.petal_width,fill_alpha = 0.5,)
show(f)
```
