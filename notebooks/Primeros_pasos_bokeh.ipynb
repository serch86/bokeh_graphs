```python
# se importan librerias de bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show
```

```python
#se crean datos normalitos
x = [0,1,2,3,4,5]
y = [5,6,7,8,9,10]
```

```python
#archivo de salido que se guardara en tu carpeta para interactuar con el grafico
output_file("line.html")
# se crea la figura como en matplotlib
f=figure()
# se rellena la figura
f.line(x,y)

f.circle(x,y)

f.triangle(x,y)

# muestra la figura creada...
show(f)
```
